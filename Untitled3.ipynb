```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression


data = pd.read_excel('boston_data for piton.xlsx')

data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.describe()
df.info()


plt.scatter(data.x, data.Y, alpha = 0.5) 

plt.xlim(10,16) 
plt.ylim(2.5,3.5) 


plt.xlabel ('X')
plt.ylabel ('Y')


x =pd.DataFrame(data.x)
y = pd.DataFrame(data.Y)
lr = LinearRegression()
lr.fit (x,y)


pred = lr.predict(x)
lr.coef_, lr.intercept_

```

```python
plt.scatter(x, y, alpha = 0.5 ) 

plt.plot(x, pred, color = 'red', linewidth = 2)

plt.xlim(10,16) 
plt.ylim(2.5,3.5) 


plt.xlabel ('X')
plt.ylabel ('Y')


"R^2 = " + str(round(lr.score(x,y),2))


```
