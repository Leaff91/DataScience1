```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
sp = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
sp= sp\
                                    .rename(columns=
                                            {'parental level of education': 'parental_level_of_education',
                                            'test preparation course': 'test_preparation_course',
                                            'math score': 'math_score',
                                            'reading score': 'reading_score',
                                            'writing score': 'writing_score'})



```

```python
sp.head(5)
sp.reading_score.hist()
#list(sp)

sp.plot.scatter(x = 'math_score', y = 'reading_score' )

```

```python
sns.lmplot(x = 'math_score', y = 'reading_score', data = sp, hue = 'gender' )
sns.lmplot(x = 'math_score', y = 'reading_score', data = sp, hue = 'test_preparation_course' )
aa = sns.lmplot(x = 'math_score', y = 'reading_score', data = sp, hue = 'lunch' )
aa.set_xlabels('Math score')
aa.set_ylabels('Reading score')
```
