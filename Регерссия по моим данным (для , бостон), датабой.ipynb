```python

x = pd.DataFrame(d.x)
y = pd.DataFrame(d.y)

lr = LinearRegression()
lr.fit(x, y) 
```

```python
pred = lr.predict(x)

import sklearn.metrics as metrics

y_true = y.y
y_pred = pred

r2 = metrics.r2_score(y_true, y_pred)
print('r2: ', round(r2, 4))
```
