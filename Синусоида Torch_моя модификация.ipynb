````python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (25.0, 7.0)
````

````python
x_train = torch.rand(1000)
x_train = x_train *10.0 - 5.0
print(max(x_train))
print(min(x_train))
````

````python
y_sub_train = torch.cos(x_train)**2
````

````python

plt.plot(x_train.numpy(), y_sub_train.numpy(),'o')
plt.title('$ y = cos^2(x)$')
````

````python
y_sub_train.shape
````

````python
#noisy = torch.randn(y_sub_train.shape)/0.5
noisy = torch.randn(y_sub_train.shape)/4
plt.plot(x_train.numpy(), noisy.numpy(),'o')
plt.title('$ noisy  $')
````

````python
y_train = y_sub_train + noisy
plt.plot(x_train.numpy(), y_train.numpy(),'o')
plt.title('$ y = cos^2(x) + noisy  $')
````

````python
x_train.unsqueeze_(1) # _ - значит меняем сам объект
y_train.unsqueeze_(1);
````

````python
x_val = torch.linspace(-5,5,100) # равномерно распределенные точки
y_val = torch.cos(x_val.data)**2
plt.plot(x_val.numpy(),y_val.numpy(), 'o')
````

````python
x_val.unsqueeze_(1)
y_val.unsqueeze_(1);
def predict(net, x, y):
  y_pred = net.forward(x)
  plt.plot(x.numpy(), y.numpy(), 'o', c='g', label ='то что должно быть')
  plt.plot(x.numpy(), y_pred.data.numpy(), 'o', c='r', label ='Предсказание сети')
  plt.legend(loc = 'upper left')
````

````python
class OurNet(torch.nn.Module):
  def __init__(self, n_hid_n):
    super(OurNet, self).__init__()
    self.fc1 = torch.nn.Linear(1, n_hid_n) #fc - full connected, Linear - Линейный полносвязный слой
    self.act1 = torch.nn.Sigmoid() # Функция активации
    #self.fc2 = torch.nn.Linear(n_hid_n, n_hid_n)
    #self.act2 = torch.nn.Sigmoid()
    self.fc3 = torch.nn.Linear(n_hid_n, 1)
    # 1 вход -> n нейронов ->n нейронов ->n нейронов -> 1 выход 
  
  def forward(self, x):
    x=self.fc1(x)
    x=self.act1(x)
    #x=self.fc2(x)
    #x=self.act2(x)
    x=self.fc3(x)
    return x
````

````python
our_net = OurNet(10) # количество нейронов

optimizer = torch.optim.Adam(our_net.parameters(), lr = 0.001) # В our_net.parameters находятся ВЕСА!
def loss(pred, true):
  sq = (pred-true)**2
  return sq.mean()
````

````python
predict (our_net, x_val, y_val)
````

````python


k=10000 #Это эпохи


for e in range(k): #Это эпохи
  optimizer.zero_grad()
  y_pred = our_net.forward(x_train) 
  loss_val = loss(y_pred, y_train)
  
  loss_val.backward()
  optimizer.step()

  if not e % 100:
    print(loss_val)

````

````python
predict(our_net, x_val, y_val)
````
