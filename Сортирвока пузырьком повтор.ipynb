```python
mas = [1,4,5,2]
n  = len(mas)
count_sravn = 0
count_peremech = 0
print(n)
for j in range(n-1):
  for i in range(n-1-j):
    if mas[i] > mas[i+1]:
       mas[i], mas[i+1] = mas[i+1], mas[i]
       count_peremech +=1
    count_sravn+=1
  print(mas)  

print('count_peremech', count_peremech)  
print('count_sravn', count_sravn)  

```
