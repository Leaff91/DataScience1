```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn
from sklearn.linear_model import LinearRegression

d = pd.read_csv("boston.csv", sep = ';', decimal = ',', header= None)
d
```

```python
d.columns = ['x', 'y']
d
```

```python
d.describe().round()
```

```python
# рисуем график рассеяния 
# 1й вариант
d.plot.scatter(x = 'x', y='y')
```

```python
# 2й вариант
plt.scatter(d.x, d.y, alpha = 0.6  )
```

```python
# построим регрессию, вывести коэффициенты
x = pd.DataFrame(d.x)
y = pd.DataFrame(d.y)

reg = LinearRegression()
reg.fit(x, y)

reg.coef_, reg.intercept_
```

```python
# построим Y_прогнозное, вывести его
pred = reg.predict(x)
pred
```

```python
# посчитаем Р квадрат 1м способом
import sklearn.metrics as metrics

y_true = y.y
y_pred = pred

r2 = metrics.r2_score(y_true, y_pred)
r2 = r2.round(4)
print(f'R квадрат = {r2}')
```

```python
# посчитаем Р квадрат 2м способом (встроено в Linear Regr)

r2 = reg.score(x, y)
r2 = r2.round(4)
print(f'R квадрат = {r2}')
```

```python
# Нарисовать и исходные точки, и регрессию красным
plt.scatter(d.x, d.y)
plt.plot(x, reg.predict(x), color = 'red')
```

```python
# Корреляции
r = d.corr()
r = r['x']['y'].round(4)
r
```
