```python
from sklearn.model_selection import cross_val_score
scores_data = pd.DataFrame()

for i in range(1, 30):
    clf = tree.DecisionTreeClassifier(criterion = 'entropy', max_depth = i)
    clf.fit(x_train, y_train)
    train_score = clf.score(x_train, y_train)
    test_score = clf.score(x_test, y_test)
    mean_cross_val_score = cross_val_score(clf, x_train, y_train , cv = 5).mean()
    temp_score_data = pd.DataFrame({'max_depth': [i], 'train_score': [train_score], 'test_score': [test_score], 'cross_val_score': [mean_cross_val_score]})
    scores_data = scores_data.append(temp_score_data)

#round(scores_data,2)
scores_data.plot(x = 'max_depth', y = ['train_score', 'test_score', 'cross_val_score'], figsize = (15, 10)) #   ,   

```
