```python
import math
 
print("Введите коэффициенты для уравнения")
print("ax^2 + bx + c = 0:")
a = float(input("a = "))
b = float(input("b = "))
c = float(input("c = "))

discr = b ** 2 - 4 * a * c

if a == 0:
    print("Это не квадратное уравнение")
else:
    if discr > 0:
        x1 = (-b + math.sqrt(discr)) / (2 * a)
        x2 = (-b - math.sqrt(discr)) / (2 * a)
        print("Уравнение имеет два корня")
        print("x1 = %.5f \nx2 = %.5f" % (x1, x2))
    elif discr == 0:
        x = -b / (2 * a)
        print("Уравнение имеет один корень")
        print("x = %.2f" % x)
    else:
        print("Корней нет")
```
