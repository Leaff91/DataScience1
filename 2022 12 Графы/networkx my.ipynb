```python
# Граф 13. Для  графа рассчитать кратчайший путь и кратчайшее расстояние 

G = nx.gnm_random_graph(n=11,m=20, seed=2)

# 1й вариант визуализации
nx.draw(G, with_labels=True, node_color = 'yellow', edge_color = 'black', width=0.2, alpha=0.9) # width=0.3 толщина линий , , alpha=0.5  - прозрачность вершин и их подписей 

# 2й вариант визуализации
pos=nx.spring_layout(G)
plt.figure(figsize=(15,10))
nx.draw(G, pos=pos, with_labels=True, width=0.2, node_color='yellow', edge_color = 'black')


nx.number_of_edges(G) #возвращает количество ребер в графе

source=0
target=8
print ('метод Дейкстры, чтобы найти кратчайший путь:')
path=nx.dijkstra_path(G, source=source, target=target)
print (path)

print ('метод Дейкстры, чтобы найти кратчайшее расстояние:')
distance = nx.dijkstra_path_length(G, source=source, target=target)
print ('Расстояние  равно:', distance)
```
