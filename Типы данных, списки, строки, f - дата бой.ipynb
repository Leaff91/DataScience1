```python
def min0(*a):
  m = a[0]
  for x in a:
    if m > x:
      m = x
  return m

min0(1,2,-100,0, -100,1,1,1)


def max0(*a):
  m = a[0]
  for x in a:
    if m < x:
      m = x
  return m

max0(1,2,3,4,6,12)
```
