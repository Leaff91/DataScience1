```python
import pandas as pd

df = pd.read_csv('atp_matches_2022.csv', sep=',')
df.head()
```

```python
df.columns
```

```python
df = df[['tourney_id', 'winner_id', 'winner_name', 'loser_id',  'loser_name', 'winner_rank', 'loser_rank' , 'winner_rank_points', 'loser_rank_points' ]]
df.head(8)
```

```python
# Задача1
# Найдем среднюю вероятность победы всех игроков в рассматриваемом массиве, затем найдем вероятность победы игроков, у которых разница в рейтинге  от  20 пунктов. 
# И сравним эти вероятности

# Турниров много (145):
df['tourney_id'].unique()
```

```python
df['tourney_id'].nunique()
```

```python
# На первом шаге рассмотрим один турнир '2022-8888'
cond0 = df['tourney_id'] == '2022-8888'
df[cond0].head()
```

```python
# Создадим новую таблицу df2, включающую и выигрыши, и проигрыши тенниссистов
# df2 = df[cond0]
df2 = df
df2 = df2.drop(['winner_rank_points', 'loser_rank_points'], axis = 1)
df2.columns = ['tourney_id', 'a_id', 'a_name', 'b_id', 'b_name', 'a_rank', 'b_rank']


df3 = df2[['tourney_id', 'b_id', 'b_name', 'a_id', 'a_name', 'b_rank', 'a_rank'] ]
df3.columns = ['tourney_id', 'a_id', 'a_name', 'b_id', 'b_name', 'a_rank', 'b_rank']

df3['win']=0
df3['lose']=1
df3


df2['win']=1
df2['lose']=0
```

```python
df4 = pd.concat([df2, df3])
df4
```

```python
# Добавим колонку с критерием разницы в ранке больше или равно 20
df4['delta'] = df4['b_rank'] - df4['a_rank'] >= 20
df4
```

```python
df4['Total'] = df4['win'] + df4['lose']
df5 = df4.groupby('a_name')[['win', 'lose', 'Total']].sum()
df5.head()

```

```python
#Найдем вероятность выигрыша в общем случае
P1 = df5.win.sum()/df5.Total.sum()
P1 = P1.round(2)
P1
```

```python
#Найдем вероятность выигрыша для delta>20

df6 = df4[df4['delta'] == 1].groupby('a_name')[['win', 'lose', 'Total']].sum()
P2 = df6.win.sum()/df6.Total.sum()
P2 = P2.round(2)
P2
```

```python
print(f'Средняя вероятность выигрыша игрока = {P1}. Вероятность выигрыша при разнице более 20 пунктов = {P2}. Таким образом, вероятность значимо повышается')
```
